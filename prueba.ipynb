### Cargada de datos de los diferentes excel

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as mp

excelTvDia1 = 'TvDia1.xlsx'
excelTvDia2 = 'TvDia2.xlsx'
excelTvDia3 = 'TvDia3.xlsx'


df_TvDia1 = pd.read_excel(excelTvDia1)
df_TvDia2 = pd.read_excel(excelTvDia2)
df_TvDia3 = pd.read_excel(excelTvDia3)

```
